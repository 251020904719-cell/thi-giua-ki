```python
n = int(input("nhập số nguyên dương: "))
tong = 0
for i in range(1, n+1):
 if i % 2 == 0:
  tong +=i
print(tong)
```

```python
s = input("nhập chuỗi: ")
words = s.split()
print("số từ trong chuỗi:", len(words))
tu_dai_nhat = words[0]
for tu in words:
   if len(tu) > len(tu_dai_nhat):
      tu_dai_nhat = tu
      print("từ dài nhất là:", tu_dai_nhat)


```

```python

ds = list(map(int, input("Nhập danh sách số nguyên: ").split()))
ds_khong_trung = sorted(set(ds))
print("Danh sách đã sắp xếp:", ds_khong_trung)
if len(ds_khong_trung) >= 2:
    print("Số lớn thứ hai:", ds_khong_trung[-2])
else:
    print("Không đủ phần tử để tìm số lớn thứ hai")
```

```python
diem = list(map(float, input("danh sách điểm: ").split()))
diem_hop_le = [d for d in diem if 0 <= d <= 10]
diem_dat = [d for d in diem_hop_le if d >= 5]
diem_tb = sum(diem_dat)/len(diem_dat) if diem_dat else 0
so_dat = len(diem_dat)
print(f"trung bình: {diem_tb:.6f}")
print(f"đạt: {so_dat}")
```
